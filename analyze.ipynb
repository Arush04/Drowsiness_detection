```python
df2.sort_values(by='DatTime', inplace=True)

# Extract 'DatTime' and 'Lane' columns
datetime = df2['DatTime']
lane = df2['Lane']

# Initialize variables
previous_lane = None
lane_changes = []

# Iterate through each row in the DataFrame
for index, current_lane in enumerate(lane):
    # Check if the current lane is different from the previous lane
    if current_lane != previous_lane and previous_lane is not None:
        lane_changes.append((datetime[index], previous_lane))

    # Update the previous lane for the next iteration
    previous_lane = current_lane

# Extract DatTime and Lane from lane_changes
lane_change_datetimes = [change[0] for change in lane_changes]
changed_lanes = [change[1] for change in lane_changes]
# Plot the points when lane changes occur
plt.figure(figsize=(10, 6))
plt.scatter(lane_change_datetimes, changed_lanes, color='red', label='Lane Changes')
plt.xlabel('DatTime')
plt.ylabel('Lane')
plt.title('Lane Changes')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print(len(lane_changes))
```

```python
lane_change_datetimes = [change[0] for change in lane_changes]

# Extract LaneOffset values at lane change DatTime points
lane_offset_at_changes = [df2.loc[df2['DatTime'] == change[0], 'LaneOffset'].iloc[0] for change in lane_changes]

# Plot the original graph
plt.figure(figsize=(10, 6))
plt.plot(df2['DatTime'], df2['LaneOffset'], label='LaneOffset')
plt.xlabel('DatTime')
plt.ylabel('LaneOffset')
plt.title('DatTime vs LaneOffset')

# Plot the lane change points
plt.scatter(lane_change_datetimes, lane_offset_at_changes, color='red', label='Lane Changes')
plt.legend()
plt.grid(True)
plt.show()
print(lane_changes)
```

```python
# Calculate distance between vehicles
df2['Distance'] = ((df2['LeadVehPosX'] - df2['XPos'])**2 + (df2['LeadVehPosY'] - df2['YPos'])**2)**0.5

# Define tailgating threshold
threshold = 1

# Check for tailgating
df2['Tailgating'] = df2['Distance'] < threshold

# Count tailgating instances
tailgating_count = df2['Tailgating'].sum()

print("Number of tailgating instances:", tailgating_count)
```
